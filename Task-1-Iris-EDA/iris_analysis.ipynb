# Iris Dataset Analysis

## Objective
The goal of this task is to explore and visualize the Iris dataset using Python libraries like pandas, matplotlib, and seaborn.

## Dataset Description
The Iris dataset contains measurements of iris flowers including:
- Sepal length
- Sepal width
- Petal length
- Petal width
- Species (Setosa, Versicolor, Virginica)

## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Dataset

```python
# Load dataset from seaborn
df = sns.load_dataset("iris")

# Display first 5 rows
df.head()
```

## 3. Basic Data Inspection

```python
print("Shape: ", df.shape)
print("Columns: ",df.columns)

df.info()
df.describe()
```

- The dataset contains 150 rows and 5 columns.
- There are no missing values in the dataset.
- All features are numerical except the species column.

## Visualizations (CORE PART)

## 4. Scatter Plot

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

- Setosa is clearly separable from the other two species.
- Versicolor and Virginica show overlap, making classification harder between them.

```python
df.groupby('species').mean()
```

- “Virginica has highest petal dimensions”
- “Setosa has smallest measurements”

## 5. Histogram (Distribution)

```python
df.hist(figsize=(10, 8))
plt.suptitle("Feature Distributions")
plt.show()
```

- The histogram shows that petal length and petal width have distinct distributions, which helps in distinguishing species more effectively than sepal features.
- Most features follow a roughly normal distribution.


## 6. Box Plot (Outliers)

```python
sns.boxplot(data=df)
plt.title("“Boxplot Analysis of Feature Distribution and Outliers”")
plt.show()
```

- Some minor outliers are visible in sepal width.
- Petal features show less variation compared to sepal features.

## 7. Pairplot

```python
sns.pairplot(df, hue='species')
plt.show()
```

**Insight:**
* Petal features are strongest for classification

## 8. Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

**Insight:**
* Petal length & width are highly correlated

## Key Insights

| Feature        | Observation |
|---------------|------------|
| Sepal Length  | Moderate variation across species |
| Sepal Width   | Some overlap and minor outliers |
| Petal Length  | Strong separation between species |
| Petal Width   | Highly correlated with petal length |

## Conclusion

- The dataset is clean and contains no missing values.
- Petal length and petal width are the most influential features for classification.
- Setosa is linearly separable, making it easy to classify.
- Versicolor and Virginica overlap, requiring more advanced models for accurate classification.
- Data visualization provides strong insights before applying machine learning models.

This analysis demonstrates how exploratory data analysis helps in understanding patterns and preparing data for machine learning models.
